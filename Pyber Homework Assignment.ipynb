## Pyber Homework Assignment

Analysis:
1. There is a significantly higher count of urban drivers and ride count than that of rural areas.
2. The average fare for urban areas are lower than the fares of suburban and rural areas, which may be from the shorter distances required to travel within urban destinations.
3. There are noticeably more outliers among rural and suburban locations, versus that of urban locations

```python
#import & merge the csv files
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#read our data as a dataframe
ride_df = pd.read_csv("raw_data/ride_data.csv")
city_df = pd.read_csv("raw_data/city_data.csv")
pyber_data = pd.merge(ride_df,city_df, how="left", on="city")
pyber_data.head()
```

## Bubble Plot

```python
#Filter by City Type (Urban, Suburban, Rural)
#iloc is an option as well
urban = pyber_data[pyber_data['type'] == 'Urban']
suburban = pyber_data[pyber_data['type'] == 'Suburban']
rural = pyber_data[pyber_data['type'] == 'Rural']

#Average Fare ($) Per City
urban_average_fare = urban.groupby("city").mean()["fare"]
suburban_average_fare = suburban.groupby("city").mean()["fare"]
rural_average_fare = rural.groupby("city").mean()["fare"]

#Total number of rides per city
urban_ride_count = urban.groupby("city").count()["ride_id"]
suburban_ride_count = suburban.groupby("city").count()["ride_id"]
rural_ride_count = rural.groupby("city").count()["ride_id"]

#total number of drivers per city
urban_driver_count = urban.groupby("city").mean()["driver_count"]
suburban_driver_count = suburban.groupby("city").mean()["driver_count"]
rural_driver_count = rural.groupby("city").mean()["driver_count"]

```

```python
#Bubble Plot

plt.scatter(urban_ride_count, urban_average_fare, s = 10*urban_driver_count, 
            marker='o', edgecolor = 'black', color = "lightcoral", linewidths=1 ,
            alpha=0.8, label="Urban")

plt.scatter(suburban_ride_count, suburban_average_fare, s = 10*suburban_driver_count, 
            marker='o',edgecolor = 'black', color = 'lightskyblue', linewidths=1 ,
            alpha=0.8, label="Suburban")

plt.scatter(rural_ride_count, rural_average_fare, s = 10*rural_driver_count, 
            marker='o', edgecolor = 'black', color = 'gold', linewidths=1 ,
            alpha=0.8, label="Rural")

plt.title("Pyber Ride Sharing Data")
plt.ylabel("Average Fare")
plt.xlabel("Ride Count")
plt.grid()
plt.legend(loc="best")
plt.show()
```

## Pie Charts

## % of Total Fares by City Type

```python
#% of Total Fares by City Type
urban_total_fare = urban.sum()["fare"]
suburban_total_fare = suburban.sum()["fare"]
rural_total_fare = rural.sum()["fare"]

labels = ["Urban", "Suburban", "Rural"]
sizes = [urban_total_fare, suburban_total_fare, rural_total_fare]
colors = ["lightcoral", "lightskyblue", "gold"]
explode = (0.1, 0, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=270)
plt.axis("equal")

plt.title("% of Fares by City Type")

plt.show()
```

```python
#% of Total Fares by City Type (Dennis' answer)

percent_by_type = pyber_data.groupby("type").sum()["fare"] / pyber_data["fare"].sum() * 100

plt.pie(percent_by_type,
        labels = ["Rural", "Suburban", "Urban"],
        colors = ["gold", "lightskyblue", "lightcoral"],
        explode = [0, 0, 0.1],
        autopct = '%1.1f%%',
        shadow=True,
        startangle = 150)
plt.title("% of Fares by City Type")
plt.show()
```
